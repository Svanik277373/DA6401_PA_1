```python
import wandb
from keras.datasets import mnist

(X_train, y_train), _ = mnist.load_data()

table = wandb.Table(columns=["label", "image"])

for digit in range(10):

    idx = np.where(y_train == digit)[0][:5]

    for i in idx:
        table.add_data(digit, wandb.Image(X_train[i]))

wandb.init(project="da6401_assignment1", name="Q_2.1")
wandb.log({"dataset_samples": table})

# wandb.finish()
```
